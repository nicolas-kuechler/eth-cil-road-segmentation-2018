```python
%matplotlib inline
from skimage import measure
from skimage import filters
from PIL import Image
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
gt_foreground_threshold = 0.02
component_max_size = 512
model_name = 'deep-residual-unet'

data_dir = './../../output/' + model_name + '/test_output/'
output_dir = './../../output/connected_components/'

files = os.listdir(data_dir)

for file in tqdm(files):
    img = Image.open(data_dir + file) 
    img = np.asarray(img)/255.0
    mask = np.where(img>gt_foreground_threshold, 1.0, 0.0)
    
    mask = eliminate_independent_components(mask, component_max_size)
    
    img = Image.fromarray((mask * 255).astype('uint8'))
    img.save(output_dir + file)
    
```

```python
def eliminate_independent_components(mask, component_max_size):
    ones = np.ones(mask.shape)

    for c in range(component_max_size):
        labels = measure.label(mask, background=0, connectivity=2)
        count_bg = 0
        count_fg = 0
        n_components = np.max(labels)+1
        for i in range(n_components):
            size = ones[labels==i].sum()
            if size < c:
                
                if i == 0:
                    count_fg += 1
                    mask[labels==i] = 1 # if background -> make to foreground
                else:
                    count_bg += 1
                    mask[labels==i] = 0 # if foreground -> make to background 
        if count_fg + count_bg > 0:
            print('Total Number of Components: ', n_components)
            print('Changed {} Components of Size {} from BG -> FG'.format(count_fg, c))
            print('Changed {} Components of Size {} from BG -> FG'.format(count_bg, c))
            
    return mask
    
```
